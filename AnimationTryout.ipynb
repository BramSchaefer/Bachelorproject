```python
# Import used packages
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import random
import networkx as nx
```

```python
def startcolor(nodes, sickPerc, radius, color):
    Network = nx.random_geometric_graph(nodes, radius)
    Situation = []
    Colorlist = []
    for i in range(nodes):
        randomN = random.uniform(0, 1)
        if randomN < sickPerc:
            Situation.append(1)
            Colorlist.append(color[1])
        else:
            Situation.append(0)
            Colorlist.append(color[0])
    return Network, Situation, Colorlist
```

```python
def updateSIRcolor(y, beta, gamma, Network, color):
    yNew = []
    Colorlist = []
    for i in range(len(Network)):
        if y[i] == 0:
            Connected = Network[i]
            Phealthy = 1
            for j in Connected:
                if y[j] == 1:
                    Phealthy *= (1 - beta)
            PSick = 1 - Phealthy
            randomN = random.uniform(0, 1)
            if randomN < PSick:
                yNew.append(1)
                Colorlist.append(color[1])
            else:
                yNew.append(y[i])
                Colorlist.append(color[y[i]])
        elif y[i] == 1:
            randomN = random.uniform(0, 1)
            if randomN < gamma:
                yNew.append(2)
                Colorlist.append(color[2])
            else:
                yNew.append(y[i])
                Colorlist.append(color[y[i]])
        else:
            yNew.append(y[i])
            Colorlist.append(color[y[i]])
    return yNew, Colorlist

def AnimatespreaddiseaseSIRcolor(NLoops, StartSituation, beta, gamma, Network, colorset, Colorlist):
    yList = [StartSituation]
    Colormatrix = [Colorlist]
    yNew = StartSituation
    pos = nx.get_node_attributes(Network, 'pos')
    for i in range(NLoops):
        y = yNew
        yNew, Colorlist = updateSIRcolor(y, beta, gamma, Network, colorset)
        yList.append(yNew)
        Colormatrix.append(Colorlist)
    return yList, Colormatrix

def AnimateeverythingSIRcolor(nodes, sickPerc, radius, NLoops, beta, gamma, color):
    Network, StartSituation, Colorlist = startcolor(nodes, sickPerc, radius, color)
    pos = nx.get_node_attributes(Network, 'pos')
    
    situations, Colormatrix = AnimatespreaddiseaseSIRcolor(NLoops, StartSituation, beta, gamma, Network, color, Colorlist)
    return Network, Colormatrix, pos
    #S, I, R = countingSIR(situations)
    #time = np.arange(0, NLoops + 1, 1)
    #Fig_Data = plt.figure()
    
    #plt.plot(time, S, label = 'S', color= color[0])
    #plt.plot(time, I, label = 'I', color= color[1])
    #plt.plot(time, R, label = 'R', color= color[2])
    
    #plt.title('', size= 20)
    #plt.ylabel('', size= 20)
    #plt.xlabel('', size= 20)
    #plt.legend(fontsize= 12)
    #plt.tick_params(axis= 'both', which= 'major', labelsize= 14)
    #plt.grid()
    #plt.axis('off')
```

```python
#Animation for SIR
nodes = 200
sickPerc = 0.1
radius = 0.1
beta = 0.2
gamma = 0.06
NLoops = 20
colorset = ['palegreen', 'tomato', 'deepskyblue']

fig, ax = plt.subplots(figsize = (8, 8))
Network, Colormatrix, pos = AnimateeverythingSIRcolor(nodes, sickPerc, radius, NLoops, beta, gamma, colorset)
def animateSIR(i):
    ax.clear()
    plt.axis('off')
    nx.draw_networkx_edges(Network, pos, alpha= 0.4, ax= ax)
    nx.draw_networkx_nodes(Network, pos, node_size= 100, node_color= Colormatrix[i], ax= ax)
    ax.set_title(i)
aniSIR = animation.FuncAnimation(fig, animateSIR, frames= NLoops + 1, repeat= True, interval= 500)
aniSIR.save('aniSIR.gif')
plt.show()
```
